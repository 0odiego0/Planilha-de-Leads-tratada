```python
import pandas as pd
```

```python
dados = pd.read_csv('leads.csv', sep=';')
dados

```

```python
colunas_para_manter = [
    'ID',
    'Etapa',
    'Nome do Lead',
    'Primeiro nome',
    'Criado',
    'Telefone de trabalho',
    'Outro número de telefone',
    'Outro e-mail',
    'Responsável',
    'Telefone [Automação]',
    'Fluxo Automação [RD Station]']
```

```python
dados_filtrados.to_excel('dados_filtrados.xlsx', index = False)

print("Arquivo salvo como 'dados_filtrados.xlsx'")
```
